## Personalized Function

```python
#auc assessment
def aucJ(true_labels, predictions):
    
    fpr, tpr, thresholds = metrics.roc_curve(true_labels, predictions, pos_label=1)
    auc = metrics.auc(fpr,tpr)

    print('auc:', auc)
    plt.plot(fpr, tpr, lw = 1)
    plt.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck') 
    plt.xlim([-0.05, 1.05])  
    plt.ylim([-0.05, 1.05]) 
    pl.title("ROC curve of %s (AUC = %.4f)" % ('19weight ANN',auc))
    pl.xlabel("False Positive Rate")
    pl.ylabel("True Positive Rate")
    plt.show()
```

```python
#随机打乱数据的顺序
def randomShuffle(X, Y):
    idx = [t for t in range(DataFrame(X).shape[0])]
    random.shuffle(idx)
    X = X[idx]
    Y = Y[idx]
    print()
    print('-' * 36)
    print('合并X数据维度：', DataFrame(X).shape)
    print('合并Y数据维度：', DataFrame(Y).shape)
    print('打乱后的数据标签：', '\n', DataFrame(Y).head())
    print('-' * 36)
    return X, Y
```

```python
#添加随机波动生成数据
def synData_noise(X):
    dim = DataFrame(X).shape
    noise_array = np.random.normal(scale = 0.01, size = (dim[0], dim[1])) + 1
    return X * noise_array
```

```python
#合成数据
#简单地复制数据
def synData(X_0, Y_0, X_1, Y_1, time):

    X_0_syn = X_0
    Y_0_syn = Y_0
    for i in range(time):
        X_0_syn = np.vstack( (X_0_syn, X_0) )
        Y_0_syn = np.hstack( (Y_0_syn, Y_0) )

    print('X生成数据维度：', DataFrame(X_0_syn).shape)
    print('Y生成数据维度：', DataFrame(Y_0_syn).shape)
    print('1标签X数据维度：', DataFrame(X_1).shape)
    print('1标签Y数据维度：', DataFrame(Y_1).shape)

    #合并数据集
    X_syn = np.vstack( (X_0_syn, X_1) )
    Y_syn = np.hstack( (Y_0_syn, Y_1) )

    print()
    print('合并X数据维度：', DataFrame(X_syn).shape)
    print('合并Y数据维度：', DataFrame(Y_syn).shape)
    print(DataFrame(Y_syn).head())

    #打乱数据的顺序
    X_syn, Y_syn = randomShuffle(X_syn, Y_syn)
    
    return X_syn, Y_syn
```

```python
def plot_embedding(data, label, title):
    x_min, x_max = np.min(data, 0), np.max(data, 0)
    data = (data - x_min) / (x_max - x_min)

    fig = plt.figure()
    ax = plt.subplot(111)
    for i in range(data.shape[0]):
        plt.text(data[i, 0], data[i, 1], str(label[i]),
                 color=plt.cm.Set1((label[i]+1) / 10.),
                 fontdict={'weight': 'bold', 'size': 9})
    plt.xticks([])
    plt.yticks([])
    plt.title(title)
    return fig
```

### Downsample
将训练集中的标签为1的样本分为16份与标签为0的样本连接

```python
X_train_0 = X_train[Y_train == 0,]
Y_train_0 = Y_train[Y_train == 0]
X_train_1 = X_train[Y_train == 1,]
Y_train_1 = Y_train[Y_train == 1]
```

```python
#随机抽样
nSample = int(len(X_train_1) / 16)
idx = random.sample(range(len(X_train_1)), nSample)

X_train_1_sap = X_train_1[idx,]
Y_train_1_sap = Y_train_1[idx]

#将随机抽取的样本与0标签的样本组合起来
X_train_dsap = np.vstack((X_train_0, X_train_1_sap))
Y_train_dsap = np.hstack((Y_train_0, Y_train_1_sap))

print('重复元素个数：', len(idx) - len(set(idx)))
print(X_train_dsap.shape)
print(Y_train_dsap.shape)

#打乱数据的顺序
X_train_dsap, Y_train_dsap = randomShuffle(X_train_dsap, Y_train_dsap)
```

```python
#生成噪音数据并将其保存到字典，同时合并数据
X_train_0_synNoise = X_train_0
Y_train_0_synNoise = Y_train_0
synNoiseData = {}
for i in range(15):
    synNoiseData['X_train_0_syn_' + str(i)] = synData_noise(X_train_0)
    X_train_0_synNoise = np.vstack( (X_train_0_synNoise, synNoiseData['X_train_0_syn_' + str(i)]) )
    Y_train_0_synNoise = np.hstack( (Y_train_0_synNoise, Y_train_0) )
    
print('X噪音数据维度：', DataFrame(X_train_0_synNoise).shape)
print('Y噪音数据维度：', DataFrame(Y_train_0_synNoise).shape)
print('1标签X数据维度：', DataFrame(X_train_1).shape)
print('1标签Y数据维度：', DataFrame(Y_train_1).shape)

#合并数据集
X_train_synNoise = np.vstack( (X_train_0_synNoise, X_train_1) )
Y_train_synNoise = np.hstack( (Y_train_0_synNoise, Y_train_1) )

print()
print('合并X数据维度：', DataFrame(X_train_synNoise).shape)
print('合并Y数据维度：', DataFrame(Y_train_synNoise).shape)
print(DataFrame(Y_train_synNoise).head())

#打乱数据的顺序
X_train_synNoise, Y_train_synNoise = randomShuffle(X_train_synNoise, Y_train_synNoise)
```

## Full-Connected Neural Network(keras)

```python
def FCNN1(input_shape):
    
    X_input = Input(input_shape)
    
    X = Dense(121, init = 'glorot_normal', activation = 'relu', name = 'fcl3')(X_input)
    X = BatchNormalization(axis = 1, name = 'bn4')(X)
    
    X = Dense(57, init = 'glorot_normal', activation = 'relu', name = 'fcl4')(X)
    X = BatchNormalization(axis = 1, name = 'bn5')(X)
    
    X = Dense(23, init = 'glorot_normal', activation = 'relu', name = 'fcl5')(X)
    X = BatchNormalization(axis = 1, name = 'bn6')(X)
    
    X = Dense(9, init = 'glorot_normal', activation = 'relu', name = 'fcl6')(X)
    X = BatchNormalization(axis = 1, name = 'bn7')(X)
    
    X = Dense(5, init = 'glorot_normal', activation = 'relu', name = 'fcl7')(X)
    X = BatchNormalization(axis = 1, name = 'bn8')(X)
    
    X = Dense(1, init = 'glorot_normal', activation = 'sigmoid', name = 'fc8')(X)
    
    model = Model(inputs = X_input, outputs = X, name = 'FCNN1')
    
    return model
```

```python
def my_auc(preds, dtrain):
    label = dtrain.get_label()
    
    preds = 1.0 / (1.0 + np.exp(-preds))
 
    fpr, tpr, thresholds = metrics.roc_curve(label, preds, pos_label=1)
    auc = metrics.auc(fpr,tpr)
    
    return 'auc', auc
```

```python
def my_loss_xgb(preds, dtrain):
    
    labels = dtrain.get_label()
    preds = 1.0 / (1.0 + np.exp(-preds))
    weight_0 = 16
    grad = -labels + weight_0 * preds + (1 - weight_0) * labels * preds
    hess = weight_0 * (1 - preds) * preds + (1 - weight_0) * labels * (1 - preds) * preds
    
    return grad, hess
```
